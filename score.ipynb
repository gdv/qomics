Read the score matrix at https://github.com/gdv/qomics/raw/main/score1.csv

```python
import pandas as pd
```

```python
d = pd.read_csv("https://github.com/gdv/qomics/raw/main/score1.csv")
d
```

```python
nucleotides = list(d.columns)[1:]
```

```python
nucleotides
```

```python
d.set_index('N', inplace = True)
d
```

```python
d.loc['C']
```

```python
d.loc['C']['A']
```

```python
score = {}
for x in nucleotides:
    score[x] = {}
    for y in nucleotides:
        score[x][y] = d.loc[x][y]
        
#score['A']['C']
score
```

```python
score = {}
for x in nucleotides:
    for y in nucleotides:
        score[(x, y)] = d.loc[x][y]
        
#score[('A', 'C')]
score
```

```python
score = {}
for x in nucleotides:
    for y in nucleotides:
        score[x + y] = d.loc[x][y]
        
#score[('A' +  'C')]
score
```

```python
d2 = pd.read_csv("https://github.com/gdv/qomics/raw/main/score2.csv")
d2
```

```python
d2.set_index('N', inplace = True)
d2
```

```python
score2 = {}
for x in nucleotides:
    score2[x] = {}
    for y in nucleotides:
        score2[x][y] = d2.loc[x][y]
        
#score['A']['C']
score2
```

```python
for x in nucleotides:
    for y in nucleotides:
        if pd.isna(score2[x][y]):
            score2[x][y] = score2[y][x]
        
#score['A']['C']
score2
```

```python
score
```

```python
for x in nucleotides:
    for y in nucleotides:
        if score[x][y] != int(score2[x][y]):
            print(score[x][y], score2[x][y])
```
